```python
import numpy as np
import datetime
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
df=pd.read_csv('data_new.csv') #source John Hopkins 

df['Last Update']=pd.to_datetime(df['Last Update'])
df=df.sort_values(by=['Country/Region','Last Update'])
df['log_con']=np.log(df['Confirmed'])
df['death_rate']=df['Deaths']/df['Confirmed']
df['death_rate']=np.where((df['Deaths']==0)|(df['Confirmed']<=2),0,df['death_rate'])
df.head(5)
```

```python
def Plot3D(data,x,y,z,size,color):
    pltData=[]
    data[size]=data[size].fillna(0)
    data[size]=np.where(data[size]<0,0,data[size])
    base = go.Scatter3d(x=data[x], y=data[y], z=data[z],
                                   mode='markers', 
                                   marker=dict(
        #size=data[size],
        color=data[color],                # set color to an array/list of desired values
        colorscale='hsv',   # choose a colorscale
  # choose a colorscale
        opacity=0.8
    ))
    pltData.append(base)
    fig = go.Figure(data=pltData)
    fig.show()
```

```python
Plot3D(df,'Country/Region','Last Update','death_rate','log_con','death_rate')
```
